```python
import pandas as pd
from sqlalchemy import create_engine

# 读取清洗后的数据
cleaned_path = 'C:/Users/许佳佳/Desktop/UK_Smart_Retail_Data_Ops/cleaned_data/cleaned_'

customers = pd.read_csv(cleaned_path + 'customers.csv')
geolocation = pd.read_csv(cleaned_path + 'geolocation.csv')
order_items = pd.read_csv(cleaned_path + 'order_items.csv')
order_payments = pd.read_csv(cleaned_path + 'order_payments.csv')
order_reviews = pd.read_csv(cleaned_path + 'order_reviews.csv')
orders = pd.read_csv(cleaned_path + 'orders.csv')
products = pd.read_csv(cleaned_path + 'products.csv')
sellers = pd.read_csv(cleaned_path + 'sellers.csv')

print("✅ 清洗后数据加载成功")
```

```python
import os

# 数据库存放路径
db_path = "C:/Users/许佳佳/Desktop/UK_Smart_Retail_Data_Ops/database/"
os.makedirs(db_path, exist_ok=True)

# 创建数据库连接（文件不存在会自动创建）
engine = create_engine('sqlite:///' + db_path + 'olist_dw.db')

print(f"✅ 数据库连接创建成功")
print('sqlite:///' + db_path + 'olist_dw.db')
```

```python
# 将所有清洗后的表写入数据库
# if_exists='replace' 表示如果表已存在就覆盖，适合开发阶段反复运行
# index=False 不把 DataFrame 的行号写进去

tables = {
    'dim_customers': customers,
    'dim_geolocation': geolocation,
    'dim_products': products,
    'dim_sellers': sellers,
    'dim_orders': orders,
    'dim_payments': order_payments,
    'dim_reviews': order_reviews,
    'fact_order_items': order_items
}

for table_name, df in tables.items():
    df.to_sql(table_name, engine, if_exists='replace', index=False)
    print(f"✅ {table_name} 写入成功，共 {len(df)} 行")

print("\n🎉 所有表写入数据库完成！")
```
